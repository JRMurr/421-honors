```python
import pandas as pd
import numpy as np
from keras.preprocessing.text import Tokenizer
from scipy.spatial.distance import cosine
```

```python
DATA_DIR = '../data/'
colnames = ['Sentence 1', 'Sentence 2', 'isSim'] 
df = pd.read_csv(DATA_DIR + 'train_snli.txt',sep="	",header=None, names=colnames)
df['Sentence 1']= df['Sentence 1'].astype(str)
df['Sentence 2']= df['Sentence 2'].astype(str)
df
```

```python
sentences = np.unique(np.append(df["Sentence 1"].values, df["Sentence 2"].values))
len(sentences)
```

Create word encording using 1 hot encoding of words in each sentence

Using keras's built in tokenizer

```python
tokenizer = Tokenizer(num_words=1500)
tokenizer.fit_on_texts(sentences)
```

```python
len(tokenizer.word_counts)
```

```python
# sequences = tokenizer.texts_to_sequences(sentences)
# one_hot_results = tokenizer.texts_to_matrix(sentences, mode='tfidf')
```

```python
sequences = tokenizer.texts_to_sequences(sentences)
```

```python
for (i,(sent, seq)) in enumerate(zip(sentences, sequences)):
    print('sent:{}\nseq:{}'.format(sent,seq))
    print('--------------')
    if i > 10:
        break
```

```python
df2 = df.head(n=10000)
A = tokenizer.texts_to_matrix(df2['Sentence 1'], mode='tfidf')
B = tokenizer.texts_to_matrix(df2['Sentence 2'], mode='tfidf')
```

```python
for index, row in df2.iterrows():
    df2.set_value(index, 'cosSim', cosine(A[index], B[index]))
```
